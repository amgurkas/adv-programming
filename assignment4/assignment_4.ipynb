```python
# Q1: Create a class called BankAccount that has four attributes: bankname, firstname,
# lastname, and balance.

# The default balance should be set to 0.

# In addition, create ...
# A method called deposit() that allows the user to make deposits into their balance.
# A method called withdrawal() that allows the user to withdraw from their balance.
# Withdrawal may not exceed the available balance. Hint: consider a conditional argument
# in your withdrawal() method.
# Use the str() method in order to display the bank name, owner name, and current
# balance.
# Make a series of deposits and withdrawals to test your class.

class BankAccount:
    def __init__(self, bankname, firstname, lastname, balance=0):
        self.bankname = bankname
        self.firstname = firstname
        self.lastname = lastname
        self.balance = balance

# defining the deposit funct
    def deposit(self,amount):
      if amount > 0:
        self.balance += amount
        print(f"Deposit of ${amount:.2f} successful. New balance: ${self.balance:.2f}")
      else:
        print("Deposit amount must be positive.")

# defining the withdrawal funct
    def withdrawal(self,amount):
      if amount <=0:
        self.balance -= amount
        print(f"Withdrawal of ${amount:.2f} successful. New balance: ${self.balance:.2f}")
      else:
        print("Withdrawal unsuccessful.")

    def __str__(self):
        return f"Bank: {self.bankname}\nOwner: {self.firstname} {self.lastname}\nCurrent Balance: ${self.balance:.2f}"


# Testing the functions:
print("--- Creating a bank account ---")
account = BankAccount("ABC Bank", "Frank", "Gallagher")
print(account)

print("\n--- Making deposits and withdrawals ---")
account.deposit(500)
account.withdrawal(100)
account.deposit(250.75)
account.withdrawal(700) # This should fail due to insufficient funds
account.withdrawal(300.75)
account.deposit(-50) # This should fail
account.withdrawal(0) # This should fail

print("\n--- Final account details ---")
print(account)

```

```python
# Q2: Create a class Box that has attributes length and width that takes values for length
# and width upon construction (instantiation via the constructor).

# In addition, createâ€¦

# A method called render() that prints out to the screen a box made with asterisks of
# length and width dimensions
# A method called invert() that switches length and width with each other
# Methods get area() and get perimeter() that return appropriate geometric calculations
# A method called double() that doubles the size of the box. Hint: Pay attention to return
# value here.
# Implement __eq__ so that two boxes can be compared using ==
# Two boxes are equal if their respective lengths and widths are identical.
# A method print dim() that prints to screen the length and width details of the box
# A method get dim() that returns a tuple containing the length and width of the box
# A method combine() that takes another box as an argument and increases the length
# and width by the dimensions of the box passed in
# A method get hypot() that finds the length of the diagonal that cuts through the middle



# Instantiate 3 boxes of dimensions 5,10 , 3,4 and 5,10 and assign to variables box1,
# box2 and box3 respectively
# Print dimension info for each using print dim()
# Evaluate if box1 == box2, and also evaluate if box1 == box3, print True or False to the
# screen accordingly
# Combine box3 into box1 (i.e. box1.combine())
# Double the size of box2
# Combine box2 into box1

# load libraries
import math

class Box():
  def __init__(self, length, width):
    self.length = length
    self.width = width

  def render(self):
        print(f"--- Box ({self.length}x{self.width}) ---")
        for _ in range(self.length): # Outer loop for rows (height)
            for _ in range(self.width): # Inner loop for columns (width)
                print("*", end="") # Print asterisk without newline
            print() # Print newline after each row
        print("--------------------")
        return f"Area: {self.length*self.width}"

  def invert(self):
        # Switch length and width
        self.length, self.width = self.width, self.length

  def get_area(self):
        return self.length * self.width

  def get_perimeter(self):
        return 2 * (self.length + self.width)

  def double(self):
        # Double the dimensions - note: modifies the box in place, no return value
        self.length *= 2
        self.width *= 2

  def __eq__(self, other):
        # Two boxes are equal if their respective lengths and widths are identical
        return self.length == other.length and self.width == other.width

  def print_dim(self):
        print(f"Length: {self.length}, Width: {self.width}")

  def get_dim(self):
        return (self.length, self.width)

  def combine(self, other):
        # Increase length and width by the dimensions of the other box
        self.length += other.length
        self.width += other.width

  def get_hypot(self):
        # Find the length of the diagonal using Pythagorean theorem
        return math.sqrt(self.length**2 + self.width**2)

# Test cases:

# 3 boxes of dimensions 5,10 , 3,4 and 5,10
box1 = Box(5, 10)
box2 = Box(3, 4)
box3 = Box(5, 10)

# Print dimension info for each
print("Initial dimensions:")
box1.print_dim()
box2.print_dim()
box3.print_dim()
print()

# Evaluate if box1 == box2, and also evaluate if box1 == box3
print(f"box1 == box2: {box1 == box2}")
print(f"box1 == box3: {box1 == box3}")
print()

# Combine box3 into box1
print("Combining box3 into box1...")
box1.combine(box3)
box1.print_dim()
print()

# Double the size of box2
print("Doubling box2...")
box2.double()
box2.print_dim()
print()

# Combine box2 into box1
print("Combining box2 into box1...")
box1.combine(box2)
box1.print_dim()
print()

# Additional demonstrations (optional)
print("Final box1 details:")
print(f"Area: {box1.get_area()}")
print(f"Perimeter: {box1.get_perimeter()}")
print(f"Diagonal: {box1.get_hypot():.2f}")
print(f"Dimensions tuple: {box1.get_dim()}")
```
